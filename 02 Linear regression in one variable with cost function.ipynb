## Problem Statement

You would like a model which can predict housing prices given the size of the house.  
Let's use the same two data points as before the previous lab- a house with 1000 square feet sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000.


| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1                 | 300                      |
| 2                  | 500                      |


```python
import numpy as np
import matplotlib.pyplot as plt
```

making use of the numpy library to perform the array operations and storing the training dataset x_train as feature and y_train as target

```python
x_train=np.array([1,2],dtype=float)
y_train=np.array([300,500],dtype=float)
print(f"input feature {x_train}")
print(f"output target {y_train}")
```

Developing a function to compute the cost for different values of w and b

```python
def compute_cost(w,b,x,y):
    m = x.shape[0]
    f_wb=np.zeros(m)
    cost_sum=0
    for i in range(m):
        f_wb[i]=w*x[i]+b
        cost=(f_wb[i]-y[i])**2
        cost_sum=cost_sum+cost
        
    total_cost=cost_sum/(2*m)
    return total_cost
```

When we use w=100 and b=100 we get the cost as 12500.00

```python
w=100
b=100
total_cost=compute_cost(w,b,x_train,y_train)
print(total_cost)
```

Using the compute_output function inorder to see the plot characteristics

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
```

```python
    
temp_f_wb=compute_model_output(x_train,w,b)
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Value')
plt.plot(x_train,temp_f_wb,c='b',label='Our Prediction')
plt.title("Housing prices")
plt.xlabel("House size (in 1000 sq fts)")
plt.ylabel("house price (in 1000's of $)")
plt.legend()
plt.show()
```

Thus we saw that when w=100 and b=100 the cost is not minimum and also is verified by the plot where we find that the curve doesn't fits the training dataset well

Now changing the values of w and b as w=200 and b=100

```python
w=200
b=100
total_cost=compute_cost(w,b,x_train,y_train)
print(total_cost)
```

```python
temp_f_wb=compute_model_output(x_train,w,b)
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Value')
plt.plot(x_train,temp_f_wb,c='b',label='Our Prediction')
plt.title("Housing prices")
plt.xlabel("House size (in 1000 sq fts)")
plt.ylabel("house price (in 1000's of $)")
plt.legend()
plt.show()
```

So when w=200 and b=100 we found the cost is equal to 0 and its the minimum possible and also its verified from the plot that it best fits the training dataset

Thus we can come to a conclusion that the values of w and b which gives the minimum cost best fits the training data set
